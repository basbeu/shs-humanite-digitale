```python
import json
import os
import bz2
import io
from bz2 import BZ2File
import pandas as pd
import numpy as np
import pickle
```

```python
input_dir = "shs-data-reduced/" # update with your path
```

```python
def read_jsonlines(bz2_file):
    text = bz2_file.read().decode('utf-8')
    for line in text.split('\n'):
        if line != '':
            yield line
```

```python
archives = os.listdir(input_dir)
```

```python

journal = []
date = []
title = []
fulltext = []

for archive in archives:
    f = BZ2File(os.path.join(input_dir, archive), 'r')
    articles = list(read_jsonlines(f))

    for art in articles:
        article_dict = json.loads(art)
        #if article_dict['type'] == 'ar':# !!! drop all NZZ articles
        journal.append(article_dict['id'][:3])
        date.append(article_dict['date'])
        title.append(article_dict['title'])
        fulltext.append(article_dict['fulltext'])

df = pd.DataFrame.from_dict(
{
    'journal': journal,
    'date': date,
    'title': title,
    'fulltext':fulltext,
})
```
